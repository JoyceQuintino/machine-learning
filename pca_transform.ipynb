```python
def pca(X, var=0.95):
    pca = PCA(var)
    l_dimension_data = pca.fit_transform(X)
    print(l_dimension_data.shape)
    approx = pca.inverse_transform(l_dimension_data)
    return approx
```

```python
def plot_image_pca(X, i):
    plt.subplot(1, 1, i)
    plt.imshow(X, cmap='gray')
    plt.xticks([])
    plt.yticks([])
```
